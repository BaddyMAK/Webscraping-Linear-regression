```python
# Import required libraries for processing data
import numpy as np 
import scipy as sc 
import pandas as pd
import math 
import glob 
```

# Dati del pdf vendemia 

```python
data_PDF=pd.read_csv('Final_data_chimi.csv', sep=';')
data_PDF.head(2)
```

```python
Non_col=['Stelle', 'Zuccheri riduttori', 'acido tartarico', 'acido malico', 'Linalolo' ]
data_PDF=data_PDF.drop(Non_col, axis=1)
data_PDF.head(2)
```

# Dati del sito 

```python
data_site=pd.read_csv('Scraped_piemont_final.csv', sep=';')
data_site.head(2)
```

```python
data_site.dtypes
data_site = data_site[data_site['Anno'].notna()]
data_site.Anno = data_site.Anno.astype(int)
data_site.head(2)
```

```python
data_site.shape[0]
```

```python
Newcol=['qualità','Titolo','Prezzo','Vitigno','Alcol','Categoria', 'DOC_DOCG','Anno' ]
data_site=data_site[Newcol]
data_site.head(2)
```

# Dati meteo 

```python
Meteo_stazioni=pd.read_csv('Stazione_Vitigno_Meteo.csv', sep=';')
Meteo_stazioni
```

```python
Meteo_stazioni.dtypes
```

```python
data_PDF.columns
```

```python
data_PDF['Um.Med (%)']=""
data_PDF['Temp Med (°C)']=""
data_PDF['Precipitazioni(mm)']=""
data_PDF['Bagnatura fogliare(min)']=""
for i in range(0, data_PDF.shape[0]): 
    for j in range(0, Meteo_stazioni.shape[0]): 
        if data_PDF['DOC_DOCG'][i]==Meteo_stazioni['DOC_DOCG'][j] : 
            #print('True1')
            if data_PDF['Anno'][i]==Meteo_stazioni['Anno'][j]: 
                if data_PDF['Vitigno'][i]==Meteo_stazioni['Vitigno'][j]: 
                #print('True2')
                        data_PDF['Um.Med (%)'][i]=Meteo_stazioni['Um.Med (%)'][j]
                        data_PDF['Temp Med (°C)'][i]=Meteo_stazioni['Temp Med (°C)'][j]
                        data_PDF['Precipitazioni(mm)'][i]=Meteo_stazioni['Precipitazioni(mm)'][j]
                        data_PDF['Bagnatura fogliare(min)'][i]=Meteo_stazioni['Bagnatura fogliare(min)'][j]
```
